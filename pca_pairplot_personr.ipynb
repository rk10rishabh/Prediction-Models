```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy import stats
import seaborn as sns

data=pd.DataFrame({
    'a':np.random.randint(1,100,size=100)
})
plt.figure(figsize=(3,3))
plt.hist(data['a'],bins=10,edgecolor='k',alpha=0.5)
plt.title('Histogram of Random numbers')
plt.show()

plt.figure(figsize=(3,3))
plt.boxplot(data['a'])
plt.title('Box Plot of Numbers')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.DataFrame({
    'time':pd.date_range(start='1/1/2023',periods=100),
    'v1':np.random.randn(100).cumsum(),
    'v2':np.random.randn(100).cumsum()
})
plt.figure(figsize=(3,3))
plt.plot(df['time'],df['v1'],label='Value 1')
plt.plot(df['time'],df['v2'],label='Value 2')
plt.title('Line plot')
plt.legend()
plt.show()
print(df)
```

```python
covariance=np.cov(df['v1'],df['v2'])[0,1]
print("Covariance : ",covariance)
```

```python
from scipy.stats import pearsonr

df2=pd.DataFrame({
    'v3':np.random.randn(100),
    'v4':np.random.randn(100),
    'c1':np.random.choice(['A','B','C'],100),
    'c2':np.random.choice(['X','Y'],100)
})
plt.figure(figsize=(3,3))
sns.scatterplot(x='v3',y='v4',data=df2)
plt.title('Scatter Plot')
plt.show()
```

```python
c,_=pearsonr(df2['v3'],df2['v4'])
print("Pearson Correlation Coeffecient : ",c)
ct=pd.crosstab(df2['c1'],df2['c2'])
print("Cross Tabulation : ",ct)
```

```python
from sklearn.decomposition import PCA
df3=pd.DataFrame({
    'f1':np.random.randn(100),
    'f2':np.random.randn(100),
    'f3':np.random.randn(100),
    'f4':np.random.randn(100)
})
sns.pairplot(df3)
plt.title("Pair Plot")
plt.show()
```

```python
pca=PCA(n_components=2)
pca_result=pca.fit_transform(df3)
df_pca=pd.DataFrame(data=pca_result,columns=['PC1','PC2'])

plt.figure(figsize=(3,3))
sns.scatterplot(x='PC1',y='PC2',data=df_pca)
plt.title('PCA Result')
plt.show()
```
